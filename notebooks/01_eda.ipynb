The following graph shows how many months the patient has left to live for each cancer type.

```python
plot_survival_months_by_cancer_type(df)
```

#### Observation

Among the cancer types in the dataset, prostate cancer patients tend to have the longest survival times, while lung cancer patients tend to have the shortest.

### 14. Survival Months by Stage

```python
plot_survival_months_by_stage(df)
```

#### Observation

### 15. Status by Stage

```python
plot_status_by_stage(df)
```

#### Observation

### 16. Key Findings

3. Breast cancer is the most common cancer type represented in the dataset.

4. Stage III is the most prevalent cancer stage among patients.

5. The majority of patients in the dataset are deceased.

6. Survival time is related to both cancer type and stage. Prostate cancer patients tend to have the longest survival times, while lung cancer patients tend to have the shortest.

7. Patients diagnosed at Stage I generally have the longest survival times, while Stage IV patients have the shortest survival times and the highest deceased rates.
